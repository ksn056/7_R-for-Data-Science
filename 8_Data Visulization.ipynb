### Histogram

```r
iris
```

```r
h1<-ggplot(data=iris)+
    aes(x=Sepal.Length,fill=Species)+
    geom_histogram()
```

```r
h1
```

```r
ggplot(data = iris) +
  aes(x = Petal.Length, fill = Species)+geom_histogram()
```

```r
ggplot(data = iris) +
  aes(x = Petal.Length, fill = Species)+geom_histogram()+
  theme_classic()
```

```r
ggplot(data=iris, aes(x=Sepal.Width))+
geom_histogram(binwidth=0.2,bins=10,color="black", aes(fill=Species))+ 
labs(x="SepalWidth",y="Frequency") + ggtitle("Histogram of Sepal Width")+ theme_dark()
```

### Bar Plot

```r
library(ggplot2)
bar <- ggplot(data=iris, aes(x=Species,fill=Species))
bar + geom_bar() + 
  xlab("Species") +  ylab("Count") + ggtitle("Bar plot of Sepal Length")
```

### Density Plot

```r
ggplot(data = iris) +
  aes(x = Petal.Length, fill = Species)+geom_density(alpha = 0.3)
```

```r
ggplot(data = iris) +
  aes(x = Petal.Length, fill = Species) + 
  geom_density( alpha = 0.3) +
  facet_wrap(~ Species, nrow = 3) 
```
